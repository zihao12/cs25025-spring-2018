```python
import scipy.io
%matplotlib inline
import matplotlib.pyplot as plt
import random
import pandas as pd
import numpy as np
import sklearn
from numpy import linalg as LA
from sklearn.preprocessing import normalize
```

```python
data = scipy.io.loadmat('/Users/ontheroad/Desktop/cs25025/hw3/IMAGES_RAW.mat')
images = data['IMAGESr']
# images is a 3D array of size [512,512,10]
# where the 10 images are of size 512 x 512
#Show the first image.
plt.imshow(images[:,:,0], cmap= 'gray')
# Function to sample image patches from the large images.
## I chaneg the function a little bit to obtain flattened repre of image
# def sample_random_square_patches(image, num, width):
#     patches =numpy.zeros([width,width,num]);
#     for k in range(num):
#         i, j = random.sample(range(image.shape[0]-width),2)
#         patches[:,:,k] = image[i:i+width,j:j+width]
#     return patches

def sample_random_square_patches(image, num, width):
    patches =np.zeros([width*width,num]);
    for k in range(num):
        i, j = random.sample(range(image.shape[0]-width),2)
        patch = image[i:i+width,j:j+width]
        patches[:,k] = patch.reshape(width*width)
    return patches
```

# SGD and other facilitating functions

```python
## select pics to use as sample pool for stochastic gradient descent
num_pic = 100
pool = sample_random_square_patches(images[:,:,0],num_pic,12)  
m = pool.shape[1]
```

```python
"""
Reference: discussed with Xinyu Wei
const_eta = 0,1,2 means eta = c, eta = c/t,eta = c/sqrt(t)
"""

def sparse_encoding_sdg(pool,L = 156,alpha = 0.0001,n_iter = 3,const_eta = 0,eta = 0.1):
    from sklearn import linear_model
    from sklearn.preprocessing import normalize
    np.random.seed(12345)
    X = np.random.rand(144,L)  
    X = normalize(X,axis = 0)
    
    Xs = []
    #Xs.append(X) ## append the original X
    diffs = []
    m = pool.shape[1]
    
    
    inspect_step = pool.shape[1]//6
    
    for t in range(n_iter):
        if const_eta == 0:
            eta = eta
        elif const_eta == 1:
            eta = eta/(t+1)
        elif const_eta == 2:
            eta = eta/np.sqrt(t+1)
            
        for i in range(m):
            y = pool[:,random.randint(1,m-1)] # randomly select a point from the pool
            #y = np.array([y])
            lasso = sklearn.linear_model.Lasso(alpha=alpha,fit_intercept =True)
            lasso.fit(X,y)
            beta = lasso.coef_.transpose()
            beta = np.array([beta])
            
            #gradient = np.transpose(y-X.dot(beta))
            y = np.array([y]).transpose()
            
            gradient = y-np.dot(X,beta.transpose())

            #print("gradient ",str(gradient.shape))
            #print("beta ",str(beta.shape))
            gradient = -1*np.dot(gradient,beta)
            #print("gradient ",str(gradient.shape))
            
            X_old = X.copy()
            X -= eta*gradient
            X = normalize(X,axis = 0)
            diff = LA.norm(X-X_old)
            diffs.append(diff)
            
            if (t*m+i % inspect_step)==0:
                Xs.append(X)
            
    return X,diffs,Xs
    
            
```

```python
# Display codebook function
def show_codebook(X):
    nrows = 13
    ncols = 12
    plt.figure(figsize=(ncols, nrows))
    for i in range(nrows*ncols):
        plt.subplot(nrows, ncols, i+1)
        image = X[:,i].reshape(12,12)
        plt.imshow(image, cmap="gray")

        plt.axis("off")
    plt.axis("off")
    plt.show()
```

# Tune parameters: alpha & eta

## first tune alpha

```python
alphas = [10**x for x in range(-4,0)]
for alpha in alphas:
    X,diffs,Xs = sparse_encoding_sdg(pool,alpha = alpha)
    #plt.plot(sparse_encoding_sdg(pool,alpha = alpha)[1])
    plt.plot(diffs)
    plt.xlabel("iterations")
    plt.ylabel("differences in X")
    plt.title("convergence with respect to alpha = "+ str(alpha))
    plt.show()
    show_codebook(X)
    
```

Comment on alpha: bigger alpha brings faster convergence, but from the generated code book, they do not really capture much useful information. By comparison, alpha = 0.0001 generates informative codebook. So next, I designate alpha as 0.0001

## Tune eta
### First try constant eta

```python
etas = [10**x for x in range(-3,2)]
for eta in etas:
    X,diffs,Xs = sparse_encoding_sdg(pool,n_iter= 10,const_eta=0,eta = eta)
    #plt.plot(sparse_encoding_sdg(pool,alpha = alpha)[1])
    plt.plot(diffs)
    plt.xlabel("iterations")
    plt.ylabel("differences in X")
    plt.title("convergence with respect to eta = "+ str(eta))
    plt.show()
    show_codebook(X)
```

```python

etas = [10**x for x in range(-3,2)]
for eta in etas:
    X,diffs,Xs = sparse_encoding_sdg(pool,n_iter= 50,const_eta=0,eta = eta)
    #plt.plot(sparse_encoding_sdg(pool,alpha = alpha)[1])
    plt.plot(diffs)
    plt.xlabel("iterations")
    plt.ylabel("differences in X")
    plt.title("convergence with respect to eta = "+ str(eta))
    plt.show()
    show_codebook(X)
```

Analysis on constant etas:<br>
First, with smaller etas, the differences of X tend to converge to 0 (smaller than 2), after around 1000 times of stochastic gradient descent, while with bigger eta differences in X converges to around 6 quickly, but keeps oscillating. <br>
Second, however, convergence does not necessarily brings good codebook. Codebook from smaller etas do not genertae inoformative codebook, even after more iterations. On the contrary, bigger etas bring better code book. From my comaprison, eta = 1 and eta = 10 brings good results. <br>
Based on this, we want to choose eta from 1 and 10 with reconstructed images below (I also construct image for eta = 0.1 as a comparison).



```python
candidate_eta1 = sparse_encoding_sdg(pool,n_iter= 10,const_eta=0,eta = 1)[0]
reconstruct_and_compare(candidate_eta1)
```

```python
candidate_eta10 = sparse_encoding_sdg(pool,n_iter= 10,const_eta=0,eta = 1)[0]
reconstruct_and_compare(candidate_eta10)
```

```python
candidate_eta01 = sparse_encoding_sdg(pool,n_iter= 10,const_eta=0,eta = 0.1)[0]
reconstruct_and_compare(candidate_eta01)
```

Further analysis on constant etas:<br>
From the comparison of the reconstructed images above, we can see that eta = 1 seems have the best result, though eta = 10 is not far behind, while eta = 0.1 is apparently performing worse. We will use eta = 1 later on. 

### Compare etas that decrease with time

```python
eta = 1
for i in range(3):
    X,diffs,Xs = sparse_encoding_sdg(pool,n_iter= 10,const_eta=i,eta = eta)
    #plt.plot(sparse_encoding_sdg(pool,alpha = alpha)[1])
    plt.plot(diffs)
    plt.xlabel("iterations")
    plt.ylabel("differences in X")
    plt.title("convergence with respect to eta = 10 and const_eta = "+ str(i))
    plt.show()
    show_codebook(X)
```

Analysis on eta decreasing with respect to time<br>
Convergence improves when we let eta decrease with time. I will further compare them with reconstructed images.  

```python
candidate_eta1_conse1 = sparse_encoding_sdg(pool,n_iter= 10,const_eta=1,eta = 1)[0]
reconstruct_and_compare(candidate_eta1_conse1)
```

```python
candidate_eta1_conse2 = sparse_encoding_sdg(pool,n_iter= 10,const_eta=2,eta = 1)[0]
reconstruct_and_compare(candidate_eta1_conse2)
```

Comment on eta:<br>
From my analysis above, we can see that the performance improves dramatically when we let eta decrease with time, both in terms of convergence, and reconstructed images. They are better able to capture details compared with constant eta. Since the iteration time is not big (only 10 in these cases), the two types of rates do not show much difference, but their reconstructed images are already quite similar to the orginal ones.
